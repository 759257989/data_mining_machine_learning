# Part 2: Ames, Iowa Housing Dataset

```python
# load dataset into a dataframe

import pandas as pd
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df = pd.read_csv(url, sep="\t")
df.head()
```

## 2.1 Imputation

```python
# dimension of the dataset
print(df.shape)
```

```python
# Identify all columns with missing values 
null_counts = df.isnull().sum()
print("Columns with missing values: \n")

miss_col = null_counts[null_counts > 0]
miss_col.reset_index()
```

```python
# fill nominal feature with string "Missing"

#monimal features fillterd
nominal_feature = df.select_dtypes(include=['object', 'category']).columns

df[nominal_feature] = df[nominal_feature].fillna('Missing')
df.head()
```

```python
# fill numerical features with linear interpolation

# filter numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# linear interpolation to numeric columns
df[numeric_columns] = df[numeric_columns].interpolate(method='linear', axis=0)
df.head()
```

## 2.2 Standardization

```python
numeric_columns
```

```python
# remove Order and PID the identification columns from numeric columns, since these don't need to be standardized 
numeric_columns = list(filter(lambda x : x not in ['Order', 'PID'], numeric_columns) )
```

```python

plt.figure(figsize=(20, 8))

sns.boxplot(data=df[numeric_columns], fill=False)

plt.title('Boxplot of Numeric Columns Before Standardization', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Standardize the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

plt.figure(figsize=(20, 8))

sns.boxplot(data=df[numeric_columns], fill=False)

plt.title('Boxplot of Numeric Columns After Standardization', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Before standardization:
Lot Area and SalePrice are scaled in the hundreds of thousands, much larger than other features.

After standardization:
All features are rescaled to have mean = 0 and standard deviation = 1. Lot Area, Low Qual Fin SF, 3Ssn Porch, Misc Val, and Pool Area stand out with the most extreme outliers.

## 2.3 Feature Selection

```python
from sklearn.feature_selection import mutual_info_regression
# Estimate mututal information

# sale price as target
y = df['SalePrice']
# remove sale price from calculation
numeric_features = [col for col in numeric_columns if col != 'SalePrice']
# calculate MI
mi = mutual_info_regression(df[numeric_features], y, n_neighbors=3, random_state=42)

# convert to series 
mi_series = pd.Series(mi, index=df[numeric_features].columns, name='MI')
# sort from high to low
mi_series = mi_series.sort_values(ascending=False)

# convert to dataframe
mi_df = mi_series.reset_index().rename(columns={'index': 'feature'})
print(mi_df.to_string())
```

Based on the mututal information between numeric features and the 'Sale Price', the top 5 features are 'Overall Qual', 'Gr Liv Area', 'Garage Area', 'Year Built', 'Total Bsmt SF'. 

For the top 5 ranked numeric features: Overall Qual, Gr Liv Area, Garage Area, Year Built, and Total Bsmt SF, I expect sales price to increase as these values increase:

Overall Qual: Higher quality ratings means better materials, finishes, that directly increase the value of a house, so prices rise with higher scores.

Gr Liv Area: More living space means a larger, more desirable home, which can make prices up.

Garage Area: Bigger garages add convenience and storage, more appealing to buyers.

Year Built: Newer homes require less maintenance and have more modern designs and systems, more recent/newer houses tend to be higher price.

Total Bsmt SF: Larger basements increase usable space adding to functional living area and market value.

These features have high mutual information because they strongly reflect a home’s size, quality, and modernity — key factors that buyers care about most and are willing to pay a premium for when purchasing a house.
